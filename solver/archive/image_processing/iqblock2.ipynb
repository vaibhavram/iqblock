```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
im = mpimg.imread("images/IMG_0387.jpg")
```

```python
(width, height) = im.shape[0:2]
increment = width//8
```

```python
grid = np.zeros((8,8,3), dtype = 'uint8')

for r in range(8):
    row = r*increment
    for c in range(8):
        column = c*increment
        subset = im[column:(column + increment), row:(row + increment),]
        grid[c, r, :] = np.median(subset, axis = (0,1))
```

```python
plt.imshow(grid)
plt.show()
```

```python
plt.imshow(im)
plt.show()
```

```python
# rotate grid
grid90 = np.rot90(grid)
plt.imshow(grid90)
plt.show()
```

```python
grid180 = np.rot90(grid, k = 2)
plt.imshow(grid180)
plt.show()
```
