```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = Image.open("images/IMG_0387.jpg")

width = im.size[0]
height = im.size[1]

increment = width//8

pixel_values = list(im.getdata())
pixel_values = np.array(pixel_values, dtype = "float32").reshape((width, height, 3))
```

```python
pix = im.load()
```

```python
pix[1500,1500]
```

```python
pixel_values[1500,1500]
```

```python
grid = np.zeros((8,8,3))
```

```python
for r in range(8):
    row = r*increment
    for c in range(8):
        column = c*increment
        subset = pixel_values[column:(column + increment), row:(row + increment),]
        grid[c, r, :] = np.median(subset, axis = (0,1))
```
